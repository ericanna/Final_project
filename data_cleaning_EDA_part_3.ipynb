```python
def fiction_nonfiction(i):
    if i in ['Mystery', 'Romance', 'Fantasy', 'Young Adult', 'Paranormal', 'Historical', 'Science Fiction', 'Fiction','Classics', 'New Adult', 'Contemporary', 'Womens Fiction', 'Horror', 'Humor', 'Sequential Art', 'Lgbt', 'Sports and Games', 'War', 'Dark', 'Childrens','Christian Fiction', 'Womens Fiction', 'Adult Fiction','Short Stories', 'Animals', 'Music', 'Poetry', 'Novels', 'Thriller', 'Anthologies', 'Magical Realism', 'Realistic Fiction', 'Suspense','Category Romance', 'Science Fiction Fantasy', 'Literary Fiction','Novella', 'Asian Literature', 'Biblical Fiction', 'Adventure', 'Literature', 'Westerns', 'Dark Fantasy','Comics', 'Drama', 'Manga']:
        return 'Fiction'
    elif i in ['Nonfiction', 'Autobiography', 'Biography', 'Self Help', 'Politics','Economics', 'Social Science', 'Writing']:
        return 'Nonfiction'
    else:
        return i
```

```python
df['fiction_nonfiction'] = list(map(fiction_nonfiction, df['fiction_nonfiction']))
```

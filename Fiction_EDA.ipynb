```python
def cleaning_polyamorous(df, col1, col2,):
    conditions1 = (df[col1] == 'Polyamorous') & (df[col2] == 'Fantasy')
    df.loc[conditions1, col1] = 'Fantasy'

    conditions2 = (df[col1] == 'Polyamorous') & (df[col2] == 'Romance')
    df.loc[conditions2, col1] = 'Romance'
```

```python
def cleaning_realistic_fiction(df, col1, col2,):
    conditions1 = (df[col1] == 'Realistic Fiction') & (df[col2] == 'Young Adult')
    df.loc[conditions1, col1] = 'Young Adult'

    conditions2 = (df[col1] == 'Realistic Fiction') & (df[col2] == 'Romance')
    df.loc[conditions2, col1] = 'Romance'
    
    conditions3 = (df[col1] == 'Realistic Fiction') & (df[col2] == 'Childrens')
    df.loc[conditions3, col1] = 'Childrens'
```

```python
def cleaning_women_fiction(df, col1, col2,):
    conditions1 = (df[col1] == 'Womens Fiction') & (df[col2] == 'Mystery')
    df.loc[conditions1, col1] = 'Mystery'

    conditions2 = (df[col1] == 'Womens Fiction') & (df[col2] == 'Romance')
    df.loc[conditions2, col1] = 'Romance'
```

```python
def cleaning_contemporary(df, col1, col2,):
    conditions1 = (df[col1] == 'Contemporary') & (df[col2] == 'Young Adult')
    df.loc[conditions1, col1] = 'Young Adult'

    conditions2 = (df[col1] == 'Contemporary') & (df[col2].isin(['Romance', 'New Adult']))
    df.loc[conditions2, col1] = 'Romance'
    
    conditions3 = (df[col1] == 'Contemporary') & (df[col2].isin(['LGBT', "Lgbt"]))
    df.loc[conditions3, col1] = 'LGBT'
    
    conditions4 = (df[col1] == 'Contemporary') & (df[col2] == 'Drama')
    df.loc[conditions4, col1] = 'Drama'
    
    conditions5 = (df[col1] == 'Contemporary') & (df[col2] == 'Short Stories')
    df.loc[conditions5, col1] = 'Short Stories'
    
    conditions6 = (df[col1] == 'Contemporary') & (df[col2] == 'Adult')
    df.loc[conditions6, col1] = 'Erotica'
```

```python
def cleaning_dark(df, col1, col2,):
    conditions1 = (df[col1] == 'Dark') & (df[col2] == 'Romance')
    df.loc[conditions1, col1] = 'Romance'

    conditions2 = (df[col1] == 'Dark') & (df[col2].isin(['Erotica', 'Adult Fiction']))
    df.loc[conditions2, col1] = 'Erotica'
    
    conditions3 = (df[col1] == 'Dark') & (df[col2] == 'Mystery')
    df.loc[conditions3, col1] = 'Thriller'
    
    conditions4 = (df[col1] == 'Dark') & (df[col2] == 'Fantasy')
    df.loc[conditions4, col1] = 'Fantasy'
    
    conditions5 = (df[col1] == 'Dark') & (df[col2] == 'Young Adult')
    df.loc[conditions5, col1] = 'Young Adult'

cleaning_dark(df3, 'genre_1', 'genre_2')
```

```python
def cleaning_fiction(df, col1, col2,):
    conditions1 = (df[col1] == 'Fiction') & (df[col2] == 'Science Fiction')
    df.loc[conditions1, col1] = 'Science Fiction'

    conditions2 = (df[col1] == 'Fiction') & (df[col2] == 'Historical')
    df.loc[conditions2, col1] = 'Historical'
    
    conditions3 = (df[col1] == 'Fiction') & (df[col2] == 'Classics')
    df.loc[conditions3, col1] = 'Classics'
    
    conditions3 = (df[col1] == 'Fiction') & (df[col2] == 'Mystery')
    df.loc[conditions3, col1] = 'Mystery'
    
    conditions4 = (df[col1] == 'Fiction') & (df[col2].isin(['Contemporary', 'Womens Fiction', 'Cultural', 'Short Stories', 'American', 'Animals', 'Psychology', 'Literary Fiction', 'See top shelves…']))
    df.loc[conditions4, col1] = 'Delete'
    
    conditions5 = (df[col1] == 'Fiction') & (df[col2] == 'Romance')
    df.loc[conditions5, col1] = 'Romance'
    
    conditions6 = (df[col1] == 'Fiction') & (df[col2] == 'Childrens')
    df.loc[conditions6, col1] = 'Childrens'
    
    conditions7 = (df[col1] == 'Fiction') & (df[col2].isin(['Fantasy', 'Magical Realism', 'Paranormal']))
    df.loc[conditions7, col1] = 'Fantasy'
    
    conditions8 = (df[col1] == 'Fiction') & (df[col2] == 'Young Adult')
    df.loc[conditions8, col1] = 'Young Adult'
    
    conditions9 = (df[col1] == 'Fiction') & (df[col2] == 'Horror')
    df.loc[conditions9, col1] = 'Horror'
    
    conditions9 = (df[col1] == 'Fiction') & (df[col2].isin(['Christian Fiction', 'Christian']))
    df.loc[conditions9, col1] = 'Christian Fiction'
    
    conditions10 = (df[col1] == 'Fiction') & (df[col2].isin(['Thriller', 'Suspense']))
    df.loc[conditions10, col1] = 'Thriller & Suspense'
    
    conditions11 = (df[col1] == 'Fiction') & (df[col2].isin(['LGBT', 'Lgbt']))
    df.loc[conditions11, col1] = 'LGBT'
    
    conditions12 = (df[col1] == 'Fiction') & (df[col2] == 'Drama')
    df.loc[conditions12, col1] = 'Drama'
    
    conditions13 = (df[col1] == 'Fiction') & (df[col2] == 'Sequential Art')
    df.loc[conditions13, col1] = 'Comics, Graphic Novels & Manga'

cleaning_fiction(df3, 'genre_1', 'genre_2')
```

```python
def period(df, col1, col2):
    df[col1] = pd.to_numeric(df[col1], errors='coerce')
    conditions1 = (df[col1] > 2000)
    df.loc[conditions1, col2] = '21st century'
    conditions2 = (df[col1] <= 2000)
    df.loc[conditions2, col2] = '20th century or before'
```

```python
def cleaning_nonfiction(df, col1, col2,):
    conditions1 = (df[col1] == 'Nonfiction') & (df[col2].isin(['Autobiography', 'Biography', 'Classics']))
    df.loc[conditions1, col1] = 'Biography & Autobiography'

    conditions2 = (df[col1] == 'Nonfiction') & (df[col2].isin(['Religion', 'Spirituality', 'Christian']))
    df.loc[conditions2, col1] = 'Religion & Spirituality'
```

```python
cleaning_nonfiction(df2, 'genre_1', 'genre_2')
```

```python
def cleaning_nonfiction(df, col1, col2,):
    conditions1 = (df[col1] == 'Nonfiction') & (df[col2].isin(['History','World War II', 'Politics']))
    df.loc[conditions1, col1] = 'History'

    conditions2 = (df[col1] == 'Nonfiction') & (df[col2] == 'Animals')
    df.loc[conditions2, col1] = 'Animals'
    
    conditions2 = (df[col1] == 'Nonfiction') & (df[col2] == 'Childrens')
    df.loc[conditions2, col1] = 'Childrens'
    
    conditions2 = (df[col1] == 'Nonfiction') & (df[col2].isin(['Self Help', 'Inspirational']))
    df.loc[conditions2, col1] = 'Self Help'
    
    conditions2 = (df[col1] == 'Nonfiction') & (df[col2] == 'Humor')
    df.loc[conditions2, col1] = 'Humor'
    
    conditions2 = (df[col1] == 'Nonfiction') & (df[col2] == 'Travel')
    df.loc[conditions2, col1] = 'Travel'
    
    conditions2 = (df[col1] == 'Nonfiction') & (df[col2].isin(['Writing', 'Language', 'Art']))
    df.loc[conditions2, col1] = 'Writing'
    
    conditions2 = (df[col1] == 'Nonfiction') & (df[col2] == 'Feminism')
    df.loc[conditions2, col1] = 'Feminism'
    
    conditions2 = (df[col1] == 'Nonfiction') & (df[col2] == 'Food and Drink')
    df.loc[conditions2, col1] = 'Food and Drink'
    
    conditions2 = (df[col1] == 'Nonfiction') & (df[col2] == 'Science')
    df.loc[conditions2, col1] = 'Science'
```

```python
cleaning_nonfiction(df2, 'genre_1', 'genre_2')
```

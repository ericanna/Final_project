```python
def cleaning_cultural(df, col1, col2, col_change):
    fiction_conditions = (df[col1] == 'Cultural') & (df[col2].isin(['Romance', 'Fiction', 'Humor']))
    df.loc[fiction_conditions, col_change] = 'Fiction'

    nonfiction_conditions = (df[col1] == 'Cultural') & (df[col2].isin(['Nonfiction', 'Autobiography', 'Travel']))
    df.loc[nonfiction_conditions, col_change] = 'Nonfiction'
```

```python
cleaning_cultural(df, 'genre_1', 'genre_2', 'fiction_nonfiction')
```
